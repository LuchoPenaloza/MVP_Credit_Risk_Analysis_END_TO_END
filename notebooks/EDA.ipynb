7- Analyzing missing data: show the percentage of missing data for each column ordered by percentage descending

```python
isna_data = pd.DataFrame(data.isna().sum())
isna_data.rename(columns = {0:'COUNT'}, inplace = True)
isna_data["%"] = round((isna_data["COUNT"] / data.shape[0])*100,2)
isna_data_sort = isna_data.sort_values("%", ascending=False)
isna_data_sort.head(20)
```

```python
isnull_data = pd.DataFrame(data.isnull().sum())
isnull_data.rename(columns = {0:'COUNT'}, inplace = True)
isnull_data["%"] = round((isnull_data["COUNT"] / data.shape[0])*100,2)
isnull_data_sort = isnull_data.sort_values("%", ascending=False)
isnull_data_sort.head(20)
```
